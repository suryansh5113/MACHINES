```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression   
from mpl_toolkits.mplot3d import Axes3D
```

```python
np.random.seed(42)

heights = np.random.normal(170, 10, 100)
ages = np.random.normal(30, 5, 100)

X = np.column_stack((heights, ages))
weights = (heights * 0.4) + (ages * 1.5) + np.random.normal(20, 5, 100)
```

```python
model = LinearRegression()
model.fit(X, weights)
```

```python
predicted_weights = model.predict(X)
```

```python
r2_score = model.score(X, weights)
print ("R2 score:", r2_score)
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(heights, ages, weights, color='blue', label='Actual Weights')
ax.scatter(heights, ages, predicted_weights, color='red', label='Predicted Weights')

ax.set_xlabel('Height (cm)')
ax.set_ylabel('Age (years)')
ax.set_zlabel('Weight (kg)')
ax.set_title('Multiple Linear Regression: Height + Age → Weight')
ax.legend()
plt.show()
```
