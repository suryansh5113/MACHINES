```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)  # Reproducibility

# Generate 100 random heights (around 170cm, std 10)
heights = np.random.normal(170, 10, 100).reshape(-1, 1)

# Generate weights based roughly on heights
weights = (heights * 0.5) + np.random.normal(30, 5, 100).reshape(-1, 1)
```

```python
model = LinearRegression()

model.fit (heights, weights)

predicted_weights = model.predict(heights)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(heights, weights, color='blue', label='Actual Weights')
plt.plot(heights, predicted_weights, color='red', label='Regression Line')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Height vs Weight Prediction')
plt.legend()
plt.grid(True)
plt.show()
```
